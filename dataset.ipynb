# Creating DataSet Using WebScraping

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

pos = 12
URL = f'https://pesdb.net/pes2022/?mode=max_level&all=1&pos={pos}&offensive_awareness=41-99&ball_control=41-99&dribbling=41-99&tight_possession=41-99&low_pass=41-99&lofted_pass=41-99&finishing=41-99&heading=41-99&set_piece_taking=41-99&curl=41-99&speed=41-99&acceleration=41-99&kicking_power=41-99&jumping=41-99&physical_contact=41-99&balance=41-99&stamina=41-99&defensive_awareness=41-99&tackling=41-99&aggression=41-99&defensive_engagement=41-99&gk_awareness=40-98&gk_catching=40-98&gk_parrying=40-98&gk_reflexes=40-98&gk_reach=40-98&overall_rating=80-100&sort=id&order=a'

result = requests.get(URL)

print(result)
content = result.text

soup = BeautifulSoup(content, 'lxml')


table = soup.find('table', class_ = 'players')

headers = table.find_all('th')

titles = []
for i in headers:
    title = i.text
    titles.append(title)

titles
```

1) Create Data Set with stats as columns


```python
df = pd.DataFrame(columns=titles)
df
```

```python
values = table.find_all('tr')
values[2].text
```

```python
players = []
for i in range(1, len(values)):
  player = values[i]
  pl = []
  for j in player:
        pl.append(j.text)
  players.append(pl)
```

```python
for x in players:
  df.loc[len(df)] = x
```

2) Loop through to extract rest of pages

```python
page =2
while True:
  url = f'https://pesdb.net/pes2022/?mode=max_level&all=1&pos={pos}&offensive_awareness=41-99&ball_control=41-99&dribbling=41-99&tight_possession=41-99&low_pass=41-99&lofted_pass=41-99&finishing=41-99&heading=41-99&set_piece_taking=41-99&curl=41-99&speed=41-99&acceleration=41-99&kicking_power=41-99&jumping=41-99&physical_contact=41-99&balance=41-99&stamina=41-99&defensive_awareness=41-99&tackling=41-99&aggression=41-99&defensive_engagement=41-99&gk_awareness=40-98&gk_catching=40-98&gk_parrying=40-98&gk_reflexes=40-98&gk_reach=40-98&overall_rating=80-100&sort=id&order=a&page={page}'
  result = requests.get(url)
  content = result.text

  soup = BeautifulSoup(content, 'lxml')


  table = soup.find('table', class_ = 'players')

  values = table.find_all('tr')
  players = []
  for i in range(1, len(values)):
    player = values[i]
    pl = []
    for j in player:
        pl.append(j.text)
    players.append(pl)

  for x in players:
      df.loc[len(df)] = x

  page += 1

  if page == 30:
    break


```
